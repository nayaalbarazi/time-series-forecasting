```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf


# Download stock data
df = yf.download('AAPL', start='2020-01-01', end='2023-12-31')


# Inspect
print(df.head(10))


# Plot closing price
plt.figure(figsize=(10,6))
plt.plot(df['Close'])
plt.title('Apple Stock Closing Prices')
plt.ylabel('Price ($)')
plt.xlabel('Date')
plt.show()


# Prepare data for ARIMA by plotting differenced series
df['close_diff'] = df['Close'].diff()
df = df.dropna()

plt.figure(figsize=(10,6))
plt.plot('Close_diff')
plt.title('Differenced Closing Prices')
plt.show()


# Check ACS and PACF
plt_acf(df['Close_diff'])
plt.show()

plt_pacf(df['Close_diff'])
plt.show()


# Fit model and print summary
model = ARIMA(df['Close'],order=(1,1,1))
model_fit = model.fit()
print(model_fit.summary())


# Forecast and plot for next 30 days
forecase = model_fit.forecast(steps=30)

plt.figure(figsize(10,6))
plt.plot(df['Close'], label='Historical')
plt.plot(pd.data_range(df.index[-1], periods=30, freq='B'), forecast, label='Forecast')
plt.title('Apple Stock Price Forecast')
plt.xlabel('Date')
plt.ylabel('Price ($)')
plt.legend()
plt.show()
```

```python
import yfinance as yf
import time
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf


# Download stock data
df = pd.read_csv(r'Downloads\AAPL_2020_2023.csv', index_col=0, parse_dates=True)

# Set frequency
df = df.asfreq('B')
print(df.head())

# Plot closing price
plt.figure(figsize=(10,6))
plt.plot(df['Close'])
plt.title('Apple Stock Closing Prices')
plt.ylabel('Price ($)')
plt.xlabel('Date')
plt.show()

# Differencing
df['close_diff'] = df['Close'].diff()
df = df.dropna()

# Plot differenced series
plt.figure(figsize=(10,6))
plt.plot(df['close_diff'])
plt.title('Differenced Closing Prices')
plt.show()

# ACF and PACF to see how old values effect new values
max_lags = min(40, len(df['close_diff']) // 2) 

plot_acf(df['close_diff'], lags=max_lags)
plt.show()
plot_pacf(df['close_diff'], lags=max_lags, method='ywm')
plt.show()

# Fit ARIMA model
# Changed order from (1,1,1) to (1,1,0) to avoid potential negative dimensions
model = ARIMA(df['Close'], order=(1,1,0))
model_fit = model.fit()
print(model_fit.summary())

# Forecast next 30 business days
forecast = model_fit.forecast(steps=30)

# Create proper business day index for forecast
# Added error handling to ensure forecast index starts after the last data point
last_date = df.index[-1]
forecast_index = pd.date_range(start=last_date + pd.Timedelta(days=1), periods=30, freq='B')

# Plot forecast
plt.figure(figsize=(10,6))
plt.plot(df['Close'], label='Historical')
plt.plot(forecast_index, forecast, label='Forecast', color='red')
plt.title('Apple Stock Price Forecast')
plt.xlabel('Date')
plt.ylabel('Price ($)')
plt.legend()
plt.show()
```
